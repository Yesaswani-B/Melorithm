```python
import pandas as pd 
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import StandardScaler
```

```python
features= ['danceability','energy','loudness','speechiness','acousticness','instrumentalness','liveness','valence','tempo']
```

```python
#standardizing the data(make sure tempo does not outweigh energy just because the values are bigger)
```

```python
ss= StandardScaler()
```

```python
x=ss.fit_transform(df[features])
```

```python
x
```

```python
model=  NearestNeighbors(n_neighbors=6,metric='cosine')
```

```python
model.fit(x)
```

```python
def reccomended_song(songname):
    # find the song in our database
    matches= df[df['track_name'].str.contains(songname,case=False,na=False)]
    if matches.empty:
        return 'Song not found, try another one!'
    idx= matches.index[0]
    #find closest neighbors 
    distances, indices=model.kneighbors([x[idx]])
    #return the recommendation as output
    recommendations= df.iloc[indices[0][1:]][['track_name','artists','album_name']]
    return recommendations
    
```
